```python
import pandas as pd
```

```python
data = pd.read_csv('/content/calories.csv')
```

```python
data
```

```python
data.info()
```

```python
data.head(15)
```

```python
# Convert datenew
data['Date New']  = data['Date New'].apply(pd.to_datetime)
```

```python
data.info()
```

```python
timedata = pd.read_csv('/content/time_day.csv')
timedata
```

```python
# Mergin the two dataset and also dropping duplicate column of timeId

combineddata = data.merge(timedata , left_on = 'TimeId' , right_on = 'timeId')
combineddata.drop('timeId' , axis = 1 , inplace = True)
```

```python
combineddata
```

```python
# get the entire data with the breakfast

databreakfast = combineddata[combineddata['time'] == 'Breakfast']
databreakfast
```

```python
# get the lunch and Calories above 250

datalunchand250above = combineddata[(combineddata['time'] == 'Lunch')  & (combineddata['Calories']>250)]
datalunchand250above
```

```python
# Sum across the Calories and t5he time

combineddata.groupby('time')['Calories'].sum()
```

```python
# Sort by Calories

combineddata.sort_values(by=['Calories'] , ascending=False)
```

```python
# Get the unique items

combineddata['Item'].unique()
```

```python
# Get the unique dates count

combineddata['Date New'].nunique()
```

```python
# Rename sveral columns in  the code
map = {'Calories': 'intake' , 'time' : 'time_of_day'}
combineddata = combineddata.rename(columns = map)
combineddata
```

```python
#12 to 13 jan % change in Calories

aggdata = combineddata.groupby('Date New')['intake'].sum().reset_index()
```

```python
aggdata
```

```python
aggdata['previousvalue'] = aggdata['intake'].shift(1)
aggdata
```

```python
# numpy function
import numpy as np

combineddata['meals'] = np.where(combineddata['time_of_day'] == 'Snack' , 'small meal' , 'big meal')
combineddata
```

```python
# Rank food for each time of day


rankdata = combineddata.groupby(['time_of_day' , 'Item'])['intake'].sum().reset_index()
```

```python
rankdata['rank'] = rankdata.groupby('time_of_day')['intake'].rank(method = 'first' , ascending = False)
rankdata
```

```python
# Plotting

import matplotlib.pyplot as plt
plotdata = combineddata.groupby('time_of_day')['intake'].sum()
plotdata
plt.bar(plotdata.index ,plotdata.values)
```
